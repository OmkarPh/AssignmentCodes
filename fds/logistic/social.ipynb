```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score,precision_score,recall_score, f1_score
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
```

```python
sc = pd.read_csv('social.csv')
sc
```

```python
mmax = MinMaxScaler()
sc[['Age', 'EstimatedSalary']] = mmax.fit_transform(sc[['Age', 'EstimatedSalary']])
# sc.sort_values('EstimatedSalary')
sc.sort_values('Age')
sc.drop('User ID', axis=1, inplace=True)
sc
```

```python
pd.get_dummies(sc, columns=['Gender', 'Purchased'], drop_first=True).replace({True: 1, False: 0})
```

```python
ohe = OneHotEncoder(sparse_output=False, drop='first')
columns_to_encode = ['Gender']
encoded_cols = ohe.fit_transform(sc[columns_to_encode])
encoded_df = pd.DataFrame(encoded_cols, columns=ohe.get_feature_names_out(columns_to_encode))
# encoded_df.value_counts()
encoded = pd.concat([sc, encoded_df], axis=1)
encoded.drop(columns=columns_to_encode, inplace=True)
encoded
```

```python
x = encoded.drop('Purchased', axis=1)
y = encoded['Purchased'].v
x, y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)

cmatrix = confusion_matrix(y_test, y_pred)
cmatrix

# percentages = cmatrix / np.sum(cmatrix)
# percentages = (cmatrix / np.sum(cmatrix, axis=1)[:, np.newaxis]).round(2) * 100
# percentages

# labels = [for j in range(cmatrix)]

#     percentages = (c_matrix / np.sum(c_matrix, axis=1)[:, np.newaxis]).round(2) * 100

# labels = [[f"{c_matrix[i, j]} ({percentages[i, j]:.2f}%)" for j in range(c_matrix.shape[1])] for i in range(c_matrix.shape[0])]
# labels = np.asarray(labels)

# sns.heatmap(c, annot=True, cmap='Blues')
```
